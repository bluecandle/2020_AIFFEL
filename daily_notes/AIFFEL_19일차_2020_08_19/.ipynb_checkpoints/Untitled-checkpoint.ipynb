Publisher 값이 결측치인 자료들을 어떻게 처리하는 것이 좋을까?

배운 방법들을 생각해보자.
1. 결측치가 있는 데이터를 제거한다
2. 결측치 채워넣는다.
    - 특정 값을 지정해줄 수 있습니다. 그러나 결측치가 많은 경우, 모두 같은 값으로 대체한다면 데이터의 분산이 실제보다 작아지는 문제가 생길 수 있습니다.

    - 평균, 중앙값 등으로 대체할 수 있습니다. 1번에서 특정 값으로 대체했을 때와 마찬가지로 결측치가 많은 경우 데이터의 분산이 실제보다 작아지는 문제가 발생할 수 있습니다.

    - 다른 데이터를 이용해 예측값으로 대체할 수 있습니다. 예를 들어 머신러닝 모델로 2020년 4월 미국의 예측값을 만들고, 이 값으로 결측치를 보완할 수 있습니다.

    - 시계열 특성을 가진 데이터의 경우 앞뒤 데이터를 통해 결측치를 대체할 수 있습니다. 예를 들어 기온을 측정하는 센서 데이터에서 결측치가 발생할 경우, 전후 데이터의 평균으로 보완할 수 있습니다.
    
흠... 근데, 예측값을 만들거나 평균, 중앙값등의 값을 지정해주기가 어려운 상황임 => 결측치를 채워넣는 방법이 마땅히 보이지 않음

## 이상치
### z-score

```python
def outlier(df, col, z):
    return df[abs(df[col] - np.mean(df[col]))/np.std(df[col])>z].index
```

```python
# 전체 판매량을 기준으로 판단
df.loc[outlier(df, 'Global_Sales', 1.5)]
```

```python
# 전체 판매량을 기준으로 판단
df.loc[outlier(df, 'Global_Sales', 2)]
```

```python
# 전체 판매량을 기준으로 판단
df.loc[outlier(df, 'Global_Sales', 3)]
```

```python
def not_outlier(df, col, z):
    return df[abs(df[col] - np.mean(df[col]))/np.std(df[col]) <= z].index
```

```python
df.loc[not_outlier(df, 'Global_Sales', 3)]
```

### IQR Method

```python
data = df['Global_Sales']
Q3, Q1 = np.percentile(data, [75 ,25])
IQR = Q3 - Q1
IQR
```

```python
data[(Q1-1.5*IQR > data)|(Q3+1.5*IQR < data)]
```

```python
def outlier2(df, col):
  data = df[col]
  Q3, Q1 = np.percentile(data, [75 ,25])
  IQR = Q3 - Q1
  print(IQR)
  return data[(Q1-1.5*IQR > data)|(Q3+1.5*IQR < data)].index
  
outlier2(df, 'Global_Sales')
```

## 정규화

### 표준화 (standardization)

```python
global_standardization = (data - data.mean())/data.std()
global_standardization
```

```python
# trade 데이터를 Standardization 기법으로 정규화합니다. 
cols = ['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']
sales_Standardization= (df[cols]-df[cols].mean())/df[cols].std()
sales_Standardization.head()
```

```python
sales_Standardization.describe()
```

### min-max scaling

```python
# 데이터 x를 min-max scaling 기법으로 정규화합니다. 
global_min_max = (data-data.min())/(data.max()-data.min())
global_min_max
```

```python
df[cols] = (df[cols]-df[cols].min())/(df[cols].max()-df[cols].min())
df.head()
```

```python
df.describe()
```

## one-hot encoding

```python
print(df['Genre'].head())
print(df['Publisher'].head())
print(df['Platform'].head())

# get_dummies를 통해 국가명 원-핫 인코딩
genre = pd.get_dummies(df['Genre'])
publisher = pd.get_dummies(df['Publisher'])
platform = pd.get_dummies(df['Platform'])
genre.head()
```

```python
df = pd.concat([df, genre,publisher,platform], axis=1)
df.head()
```

```python
df.drop(['Genre','Platform','Publisher'], axis=1, inplace=True)
df.head()
```
